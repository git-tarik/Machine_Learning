```python
import requests
import pandas as pd

# create an empty list to hold temporary dataframes
dfs = []

for i in range(1, 429):
    response = requests.get(
        'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i)
    )
    temp_df = pd.DataFrame(response.json()['results'])[
        ['id','title','overview','release_date','popularity','vote_average','vote_count']
    ]
    dfs.append(temp_df)   # collect each page’s dataframe

# merge all into your final dataframe
df = pd.concat(dfs, ignore_index=True)
```
